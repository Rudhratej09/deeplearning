```python
z=x_reshape.view(1,9)
```

```python
z,z.shape
```

```python
x_stacked=torch.stack([x,x,x,x,x])
```

```python
x_stacked,x_stacked.size
```

```python
#TORCH.SQUEEZE()removesall the single dimensions from te shape of a target tensor
```

```python
x,x.shape
```

```python
x_reshapen=x.reshape(1,9)
```

```python
print(f'previous tensor {x_reshapen}')
print(f'its size {x_reshapen.shape}')
x_squeezed=x_reshapen.squeeze()
print(f'squeezed tensor {x_squeezed}')
print(f'its size after ssqueeze {x_squeezed.shape}')
```

```python
x_unsqueezed=x_squeezed.unsqueeze(dim=1)
```

```python
print(f'unsqueezed tensor {x_unsqueezed}')
print(f'its size after unssqueeze {x_unsqueezed.shape}')
```

## random seeds

```python
rand_seed=42
```

```python
torch.manual_seed(rand_seed)
random_tensor_A=torch.rand(3,3)
random_tensor_B=torch.rand(3,3)
print(random_tensor_A)
print(random_tensor_B)
print(random_tensor_A==random_tensor_B)
```

```python
torch.manual_seed(rand_seed)
random_tensor_A=torch.rand(3,3)
torch.manual_seed(rand_seed)
random_tensor_B=torch.rand(3,3)
print(random_tensor_A)
print(random_tensor_B)
print(random_tensor_A==random_tensor_B)
```
